#### Encoding

##### Label Encoding

```python
#take a copy from the original data
data_copy1 = data.copy()
```

```python
#we will need to replace each class with a label
# Iris-setosa >>> 0 
# Iris-versicolor >>> 1
# Iris-virginica  >>> 2
```

```python
#getting the classes names
class_names = data_copy1['Species'].unique()
class_names
```

```python
#using conditional indexing
data_copy1.loc[ data_copy1['Species'] == class_names[0],'Species']=0
data_copy1
```

```python
data_copy1.loc[ data_copy1['Species'] == class_names[1],'Species']=1
data_copy1
```

```python
data_copy1.loc[ data_copy1['Species'] == class_names[2],'Species']=2
data_copy1
```

```python
#take another copy from the original data
data_copy1 = data.copy()



# another approach (which makes more sense since it is faster)
for i in range(len(class_names)):
    data_copy1.loc[ data_copy1['Species'] == class_names[i],'Species']=i

data_copy1
```

#### Scaling and Normalizing

##### Scaling

```python
#get the mean of each feature

data_copy1['SepalLengthCm'].mean()
```

```python
mu1=data_copy1['SepalLengthCm'].mean()
mu2=data_copy1['SepalWidthCm'].mean()
mu3=data_copy1['PetalLengthCm'].mean()
mu4=data_copy1['PetalWidthCm'].mean()
```

```python
#get the std of each feature
data_copy1['SepalLengthCm'].std()
```

```python
std1=data_copy1['SepalLengthCm'].std()
std2=data_copy1['SepalWidthCm'].std()
std3=data_copy1['PetalLengthCm'].std()
std4=data_copy1['PetalWidthCm'].std()
```

```python
#scale each feature by sybtacting the mean and divide by std
data_copy1['SepalLengthCm']= (data_copy1['SepalLengthCm']-mu1)/std1
```

```python
data_copy1['SepalWidthCm']= (data_copy1['SepalWidthCm']-mu2)/std2
data_copy1['PetalLengthCm']= (data_copy1['PetalLengthCm']-mu3)/std3
data_copy1['PetalWidthCm']= (data_copy1['PetalWidthCm']-mu4)/std4
```

```python
data_copy1
```

```python
#now the mean is approx zero
data_copy1['SepalLengthCm'].mean() 
```

```python
#and the std is almost 1
data_copy1['SepalLengthCm'].std()
```

##### Normalizing

```python
#get a copy of original data
data_copy3 = data.copy()
```

```python
feature_names=data_copy3.columns[:-1]
```

```python
#get the min of each feature
data_copy3['SepalLengthCm'].min()
```

```python
mins= []
for feature in feature_names:
    mins.append(data_copy3[feature].min())
mins
```

```python
#get the max of each feature
data_copy3['SepalLengthCm'].max()
```

```python
maxs= []
for feature in feature_names:
    maxs.append(data_copy3[feature].max())
maxs
```

```python
#subtract each feature from the min and divide by (max-min)
i=0
for feature in feature_names:
    data_copy3[feature]= (data_copy3[feature]-mins[i])/(maxs[i]-mins[i])
    i+=1
```

```python
#all values are from 0 to 1
data_copy3
```

#### Dealing With Unknown Values

```python
data_copy4= data.copy()
```

```python
#replace with mean
data_copy4.fillna(data_copy4.mean(),inplace=True)
data_copy4
```

```python
#replace with zeros
data_copy4= data.copy()

data_copy4.fillna(data_copy4.mean(),inplace=True)
data_copy4
```

```python
#delete all rows contains nan
data_copy4= data.copy()

data_copy4.dropna(inplace=True)
data_copy4 #notice there are 110 rows only instead of 132
```
